## The Data

```python
# DATA FROM PUBLIC RECORDS (do not modify these lists)

# Type of each incident (40 total)
incident_types = [
    "theft", "assault", "theft", "burglary", "theft", "assault", 
    "vandalism", "theft", "assault", "theft", "burglary", "assault", 
    "theft", "vandalism", "robbery", "theft", "assault", "burglary",
    "theft", "vandalism", "assault", "theft", "theft", "burglary",
    "assault", "theft", "robbery", "vandalism", "theft", "assault",
    "burglary", "theft", "assault", "theft", "vandalism", "assault",
    "theft", "burglary", "assault", "theft"
]

# Response time in minutes for each incident (40 total)
response_times = [
    8.2, 6.5, 12.1, 7.3, 9.8, 14.2, 5.5, 8.9, 11.3, 6.8,
    9.2, 13.5, 7.1, 8.8, 10.2, 9.5, 7.7, 11.8, 6.3, 9.9,
    8.5, 12.3, 7.9, 10.1, 9.3, 8.1, 14.5, 7.2, 9.6, 11.2,
    6.9, 8.7, 13.1, 7.5, 9.1, 10.8, 8.3, 7.8, 12.7, 9.4
]

# Day of week when each incident occurred (40 total)
day_of_incident = [
    "Mon", "Mon", "Tue", "Wed", "Thu", "Thu", "Fri", "Fri", "Sat", "Sat",
    "Sun", "Sun", "Mon", "Mon", "Tue", "Wed", "Wed", "Thu", "Fri", "Fri",
    "Sat", "Sat", "Sun", "Sun", "Mon", "Tue", "Tue", "Wed", "Thu", "Thu",
    "Fri", "Fri", "Sat", "Sat", "Sun", "Mon", "Tue", "Wed", "Thu", "Fri"
]
```

## Part 1: Basic Calculations with Variables and Lists - SOLUTIONS

### Task 1.1: Calculate total number of incidents - SOLUTION

```python
# SOLUTION
# Use len() to count items in the list
total_incidents = len(incident_types)

print(f"Total incidents: {total_incidents}")
```

### Task 1.2: Find the fastest and slowest response times - SOLUTION

```python
# SOLUTION
# Use min() and max() built-in functions
fastest_response = min(response_times)
slowest_response = max(response_times)

print(f"Fastest response: {fastest_response} minutes")
print(f"Slowest response: {slowest_response} minutes")
```

### Task 1.3: Calculate average response time - SOLUTION

```python
# SOLUTION
# Use sum() to add all values, then divide by count
total_response_time = sum(response_times)
average_response = total_response_time / len(response_times)

print(f"Average response time: {average_response:.2f} minutes")
```

### Task 1.4: Check the first and last incidents - SOLUTION

```python
# SOLUTION
# Use list indexing - [0] for first, [-1] for last
first_incident_type = incident_types[0]
last_incident_type = incident_types[-1]

first_response_time = response_times[0]
last_response_time = response_times[-1]

print(f"First incident: {first_incident_type}, responded in {first_response_time} minutes")
print(f"Last incident: {last_incident_type}, responded in {last_response_time} minutes")
```

## Part 2: Using Loops to Count - SOLUTIONS

### Task 2.1: Count thefts - SOLUTION

```python
# SOLUTION
# Loop through each incident and count thefts
theft_count = 0

for incident in incident_types:
    if incident == "theft":
        theft_count = theft_count + 1

print(f"Number of thefts: {theft_count}")
```

### Task 2.2: Count thefts AND assaults - SOLUTION

```python
# SOLUTION
# Use separate counters and check for each type
theft_count = 0
assault_count = 0

for incident in incident_types:
    if incident == "theft":
        theft_count = theft_count + 1
    elif incident == "assault":
        assault_count = assault_count + 1

print(f"Thefts: {theft_count}")
print(f"Assaults: {assault_count}")
```

### Task 2.3: Count all property crimes vs violent crimes - SOLUTION

```python
# SOLUTION
# Property crimes: theft, burglary, vandalism
# Violent crimes: assault, robbery
property_crimes = 0
violent_crimes = 0

for incident in incident_types:
    if incident in ["theft", "burglary", "vandalism"]:
        property_crimes = property_crimes + 1
    elif incident in ["assault", "robbery"]:
        violent_crimes = violent_crimes + 1

print(f"Property crimes: {property_crimes}")
print(f"Violent crimes: {violent_crimes}")
```

## Part 3: Loops with Conditionals and Lists - SOLUTIONS

### Task 3.1: Find slow response times - SOLUTION

```python
# SOLUTION
# Build a new list with only slow responses (> 10 minutes)
slow_responses = []

for time in response_times:
    if time > 10:
        slow_responses.append(time)

print(f"Slow responses: {slow_responses}")
print(f"Number of slow responses: {len(slow_responses)}")
```

### Task 3.2: Calculate average using accumulator pattern - SOLUTION

```python
# SOLUTION
# Accumulate the total using a loop
total_time = 0

for time in response_times:
    total_time = total_time + time

# Calculate average
avg_time = total_time / len(response_times)
print(f"Average response time: {avg_time:.1f} minutes")
```

### Task 3.3: Find which incident types had slow responses - SOLUTION

```python
# SOLUTION
# Use range(len()) to access both lists at same index
slow_incident_types = []

for i in range(len(response_times)):
    if response_times[i] > 10:
        slow_incident_types.append(incident_types[i])

print(f"Types of incidents with slow response: {slow_incident_types}")
```

### Task 3.4: Count incidents by day of week - SOLUTION

```python
# SOLUTION
# Count weekends (Sat/Sun) vs weekdays
weekend_count = 0
weekday_count = 0

for day in day_of_incident:
    if day in ["Sat", "Sun"]:
        weekend_count = weekend_count + 1
    else:
        weekday_count = weekday_count + 1

print(f"Weekend incidents: {weekend_count}")
print(f"Weekday incidents: {weekday_count}")
```

## Part 4: Writing Functions - SOLUTIONS

### Task 4.1: Function to calculate percentage - SOLUTION

```python
# SOLUTION
def calculate_percentage(part, total):
    """
    Calculate what percentage 'part' is of 'total'
    
    Parameters:
    - part: the portion amount
    - total: the total amount
    
    Returns:
    - The percentage
    """
    percentage = (part / total) * 100
    return percentage


# Test the function
theft_percentage = calculate_percentage(theft_count, len(incident_types))
print(f"Thefts are {theft_percentage:.1f}% of all incidents")
```

### Task 4.2: Function to count specific incident type - SOLUTION

```python
# SOLUTION
def count_incident_type(incidents_list, target_type):
    """
    Count how many times target_type appears in incidents_list
    
    Parameters:
    - incidents_list: list of incident types
    - target_type: the type to count (like "theft" or "assault")
    
    Returns:
    - The count as an integer
    """
    count = 0
    for incident in incidents_list:
        if incident == target_type:
            count = count + 1
    return count


# Test the function
robbery_count = count_incident_type(incident_types, "robbery")
vandalism_count = count_incident_type(incident_types, "vandalism")
print(f"Robberies: {robbery_count}")
print(f"Vandalism: {vandalism_count}")
```

## Part 5 (Extra Credit): Summary Report - SOLUTION

```python
# SOLUTION
print("=" * 50)
print("COMMUNITY SAFETY ANALYSIS SUMMARY")
print("=" * 50)
print()

# Calculate additional statistics needed for summary
slow_response_count = len(slow_responses)
slow_response_pct = calculate_percentage(slow_response_count, total_incidents)

# 5 key findings:
print(f"1. Total incidents analyzed: {total_incidents}")
print()

print(f"2. Most common incident types:")
print(f"   - Theft: {theft_count} incidents ({calculate_percentage(theft_count, total_incidents):.1f}%)")
print(f"   - Assault: {assault_count} incidents ({calculate_percentage(assault_count, total_incidents):.1f}%)")
print()

print(f"3. Response time analysis:")
print(f"   - Average response time: {average_response:.1f} minutes")
if average_response > 10:
    print(f"   - WARNING: Average exceeds 10-minute target")
else:
    print(f"   - Average is within 10-minute target")
print()

print(f"4. Temporal patterns:")
if weekend_count > weekday_count:
    print(f"   - Weekends had more incidents ({weekend_count} vs {weekday_count})")
else:
    print(f"   - Weekdays had more incidents ({weekday_count} vs {weekend_count})")
print()

print(f"5. Slow response analysis:")
print(f"   - {slow_response_pct:.1f}% of responses took over 10 minutes")
print(f"   - {slow_response_count} out of {total_incidents} incidents")
print()

print("=" * 50)
print("RECOMMENDATION: Community organization should focus")
print("advocacy on reducing response times and addressing")
if property_crimes > violent_crimes:
    print(f"property crimes, which account for {property_crimes} incidents.")
else:
    print(f"violent crimes, which account for {violent_crimes} incidents.")
print("=" * 50)
```
